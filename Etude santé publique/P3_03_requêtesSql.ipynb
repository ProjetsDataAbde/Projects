### Question 19 : Écrivez les requêtes SQL permettant de connaître…
###### Les 10 pays ayant le plus haut ratio disponibilité alimentaire/habitant en termes de protéines (en kg) par habitant, puis en termes de kcal par habitant.

```python
# requête pour le ratio en disponibilité alimentaire/hab en proteine 
query='''SELECT 
    pays,
    sum(dispo_prot/1000) AS dispo_prot_kg_hab
FROM dispo_alim
GROUP BY pays
ORDER BY dispo_prot_kg_hab DESC
LIMIT 10'''
dispoProtKgHab=pd.read_sql_query(query,cnn)
dispoProtKgHab
```

```python
# requête pour le ratio en disponibilité alimentaire kcal/hab
query='''SELECT 
    pays,
    sum(dispo_alim_kcal_p_j) as dispo_alim_kcal_hab
FROM dispo_alim
GROUP BY pays
ORDER BY dispo_alim_kcal_hab DESC
LIMIT 10'''
dispoAlimkcalHab=pd.read_sql_query(query,cnn)
dispoAlimkcalHab
```

###### Pour l'année 2013, les 10 pays ayant le plus faible ratio disponibilité alimentaire/habitant en termes de protéines (en kg) par habitant.

```python
# requête pour le ratio disponibilité alim par habitant en terme de kg/proteine en 2013
query='''SELECT 
    pays,
    sum(dispo_prot/1000) AS dispo_prot_kg_hab
FROM dispo_alim
WHERE annee='2013'
GROUP BY pays
ORDER BY dispo_prot_kg_hab ASC
LIMIT 10'''
dispoProtkgHab2013=pd.read_sql_query(query,cnn)
dispoProtkgHab2013
```

###### La quantité totale (en kg) de produits perdus par pays en 2013.

```python
# requête pour la quantité totale de produits perdus en Kg par pays en 2013
query='''SELECT
     pays,sum (pertes*1e6) AS pertes_Kg
FROM equilibre_prod
WHERE annee='2013'
GROUP BY pays'''
pertesKg=pd.read_sql_query(query,cnn)
pertesKg
```

###### Les 10 pays pour lesquels la proportion de personnes sous-alimentées est la plus forte.

```python
# requête pour les 10 pays pour lesquels la proportion de personnes sous-alimentées est la plus forte
query='''SELECT 
    pays,
    AVG(nb_personnes/population*100) as le_Plus_pers_sous_alim
FROM population
    NATURAL JOIN sous_nutrition
GROUP BY pays
ORDER BY le_Plus_pers_sous_alim DESC
LIMIT 10'''
lePlusPersSousAlim=pd.read_sql_query(query,cnn)
lePlusPersSousAlim
```

###### Les 10 produits pour lesquels le ratio Autres utilisations/Disponibilité intérieure est le plus élevé

```python
# requête pour les 10 produits pour lesquels le ratio utilisation/dispo_int est le plus elevé
query='''SELECT 
    produit,
    AVG (autres_utilisations/dispo_int) as ratio_autres_Util
FROM equilibre_prod
GROUP BY produit
ORDER BY ratio_autres_Util DESC
LIMIT 10'''
ratioAutresUtil=pd.read_sql_query(query,cnn)
ratioAutresUtil
```
