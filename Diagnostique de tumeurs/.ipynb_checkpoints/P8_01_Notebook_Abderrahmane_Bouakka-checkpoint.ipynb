```python
# Fonction pour automatiser les étapes d'entrainement, de prédiction et calcul la métrique accuracy
def calculate_accuracy(classifier, X_train, X_test, y_train, y_test):
    
    """ 
    La fonction permet de réaliser les étapes d'entrainement, de prédiction 
    et de calcul de la métrique accuracy pour différents modèles de machine learning 
    """

    classifier.fit(X_train, y_train) # Entrainement du modèle.uniquement sur (X_train)
    y_pred_train = classifier.predict(X_train) # Prédiction dataset train.
    y_pred_test = classifier.predict(X_test)# Prédiction dataset test.
    accuracy_train = metrics.accuracy_score(y_train, y_pred_train)# Performance du modèle estimée par métrique accuracy dataset train
    accuracy_test = metrics.accuracy_score(y_test, y_pred_test)# Performance du modèle estimée par métrique accuracy dataset test
    print('Train accuracy:', '{:.2f}'.format(accuracy_train), 'Test accuracy:', '{:.2f}'.format(accuracy_test))# affichage de Accuracy (justesse)
    return accuracy_train, accuracy_test, classifier
```

### 3.Modélisation avec une validation croisée stratifiée

<div class ="alert alert-block alert-info"> 
    
**Infos :**
    
On va faire une validation croisée et utiliser l’intégralité de nos données pour entraîner et pour tester nos Classifiers. 

Aussi, On va effectuer un échantillonnage stratifié pour produire des plis qui contiennent un ratio représentatif de chaque classe. 

Ensuite, on va:

- Entrainer les classifiers
- Calculer la metrique accuracy
- Visualiser les Courbes ROC et matrices de confusion
- Effectuer une prédictions générales de l'AUC 
- Calculer la fonction Log LOSS
- Editer un rapport des metriques
    
</div>

[Haut de page](#sommaire)

```python
# importation table
data_2 = pd.read_csv('breast_cancer.csv', sep=',', index_col='id')

# Suppression colonne unnamed : 32
data_2.drop('Unnamed: 32', axis=1, inplace=True)

# Séparations des étiquettes et des données caractéristiques
y = data_2['diagnosis']
X = data_2.select_dtypes('number')

# Crosse validation
cross_validation = StratifiedKFold(n_splits=3, random_state=random_state, shuffle=True)

for train_index, test_index in cross_validation.split(X, y):
    X_train = X.iloc[train_index]
    X_test = X.iloc[test_index]
    print('Train', X_train.shape, 'Test', X_test.shape)
```

<a id='TC'></a>
### Entrainement des Classifiers

[Haut de page](#sommaire)

```python
# Random Forest Classifier
clsf_RdF_cv = RandomForestClassifier(n_estimators=10, random_state=random_state)
# fit model
clsf_RdF_cv.fit(X, y)
# make class predictions for the validation set
y_pred_class_RdF_cv = cross_val_predict(clsf_RdF_cv, X, y, cv = cross_validation)
# predicted probabilities for class 1
y_pred_prob_RdF_cv = cross_val_predict(clsf_RdF_cv, X, y, cv = cross_validation, method="predict_proba")
y_pred_prob_RdF_cv_class1 = y_pred_prob_RdF_cv[:, 1]


# Logistic Regression
clsf_Rl_cv = LogisticRegression(random_state=random_state)
# fit model
clsf_Rl_cv.fit(X, y)
# Make class predictions for the validation set.
y_pred_class_Rl_cv = cross_val_predict(clsf_Rl_cv , X, y, cv = cross_validation)
# predicted probabilities for class 1, probabilities of positive class
y_pred_prob_Rl_cv = cross_val_predict(clsf_Rl_cv , X, y, cv = cross_validation, method="predict_proba")
y_pred_prob_Rl_cv_class1 = y_pred_prob_Rl_cv[:, 1]

# SVM Classifier
clsf_SvM_cv = LinearSVC( random_state=random_state)
# fit model
clsf_SvM_cv.fit(X, y)
# make class predictions for the validation set
y_pred_class_SvM = cross_val_predict(clsf_SvM_cv, X, y, cv = cross_validation)
# predicted probabilities for class 1
y_pred_prob_SvM = cross_val_predict(clsf_SvM_cv, X, y, cv = cross_validation, method="decision_function")


# XGBoost Classifier
clsf_XgB_cv = XGBClassifier(n_estimators=10, random_state=random_state, eval_metric='error')
# fit model
clsf_XgB_cv.fit(X, y)
# make class predictions for the validation set
y_pred_class_XgB_cv = cross_val_predict(clsf_XgB_cv, X, y, cv = cross_validation)
# predicted probabilities for class 1
y_pred_prob_XgB_cv = cross_val_predict(clsf_XgB_cv, X, y, cv = cross_validation, method="predict_proba")
y_pred_prob_XgB_cv_class1 = y_pred_prob_XgB_cv[:, 1]
```

<a id='ACC'></a>
### Accuracy des Classifiers

<div class ="alert alert-block alert-info"> 
    
L'exactitude de la classification est le rapport entre le nombre de prédictions correctes et le nombre total d'échantillons d'entrée.
    
$ \ $   

$$Accuracy = \frac{Number\ of\ correct\ predictions}{Total\ number\ of\ predictions\ made} = \frac{TP + TN}{TP + TN + FP + FN}$$
    
 </div>

[Haut de page](#sommaire)

```python
# Précision (accuracy) Crosse validation score stratifiée
acc_RdF = cross_val_score(clsf_RdF_cv, X, y, cv=cross_validation).mean()
acc_Rl = cross_val_score(clsf_Rl_cv, X, y, cv=cross_validation).mean()
acc_SvM = cross_val_score(clsf_SvM_cv, X, y, cv=cross_validation).mean()
acc_XgB = cross_val_score(clsf_XgB_cv, X, y, cv=cross_validation).mean()

print('Mean test accuracy:',acc_RdF, acc_Rl, acc_SvM, acc_XgB)
```

<div class ="alert alert-block alert-success"> 

Le meilleur résultat de précision (accuracy) est celui de la RandomForest. Néanmoins, la précision de la classification peut nous donner le sentiment erroné d'atteindre une précision élevée, et notamment car cette métrique n'indique pas la distribution sous-jacente des valeurs de réponse. Aussi, on va mesurer d'autres métriques pour confirmer ou infirmer nos résultats.

</div>

<a id='LOSS'></a>
### Application d'une fonction de perte Log LOSS

<div class ="alert alert-block alert-info"> 

La fonction de perte va nous permettre d'évaluer l'incertitude de nos Classifiers. C'est-à-dire que plus le logloss est petit, meilleur est le modèle. En somme, plus l'incertitude est petite, meilleur est le modèle.

$ \ $

$$Log\ Loss = \frac{-1}{N} \sum_{i=1}^{N} \sum_{i=1}^{M}  y_{ij} * \log(\hat{y_{ij}})$$

$ \ $

- $y_{ij}$, indique si l'échantillon i appartient à la classe j ou non

- $p_{ij}$, indique la probabilité que l'échantillon i appartienne à la classe j

</div>

[Haut de page](#sommaire)

```python
# aplication de la fonctionà chaque classifier
loss_RdF = cross_val_score(clsf_RdF_cv , X, y, cv = cross_validation, scoring = 'neg_log_loss').mean()
loss_Rl = cross_val_score(clsf_Rl_cv, X, y, cv = cross_validation, scoring = 'neg_log_loss').mean()
loss_SvM = cross_val_score(clsf_SvM_cv, X, y, cv = cross_validation, scoring = 'neg_log_loss').mean()
loss_XgB = cross_val_score(clsf_XgB_cv, X, y, cv = cross_validation, scoring = 'neg_log_loss').mean()

print(loss_RdF, loss_Rl, loss_SvM, loss_XgB)
```

<div class ="alert alert-block alert-warning">
    
SVM Classifier ne supporte pas les estimations de probabilité. Aussi, on va définir SVMClassifier comme estimateur de base dans le CalibratedClassifierCV de Scikit-learn, qui générera des estimations de probabilité.
    
</div>

```python
# importation module CalibratedClassifierCV
from sklearn.calibration import CalibratedClassifierCV
```

<a id='ROCM'></a>
### Courbes ROC et matrices de confusion

<div class ="alert alert-block alert-info">

La courbe ROC peut être décomposée en sensibilité et spécificité. Le choix du meilleur modèle est une sorte d'équilibre entre la prédiction précise de M ou de B.
    
</div>

[Haut de page](#sommaire)

```python
# Courbe ROC et matrice de confusion RandomForest
matrix_1=metrics.plot_confusion_matrix(clsf_RdF, X, y)
plt.title('Matrice de confusion Random Forest (cv)', fontsize = 15)
plt.savefig('Graph/Matrice de confusion Random Forest (cv).png', dpi=300, transparent=True)

roc1=metrics.plot_roc_curve(estimator=clsf_RdF_cv, X=X, y=y)
plt.title('Courbe ROC Random Forest(cv)', fontsize = 15)
plt.savefig('Graph/Courbe ROC Random Forest (cv).png', dpi=300, transparent=True)


# Courbe ROC et matrice de confusion Régression Logistique
matrix_2=metrics.plot_confusion_matrix(clsf_Rl_cv, X, y)
plt.title('Matrice de confusion Régression logistique (cv)', fontsize = 15)
plt.savefig('Graph/Matrice de confusion Régression logistique(cv).png', dpi=300, transparent=True)

roc2=metrics.plot_roc_curve(estimator=clsf_Rl_cv, X=X, y=y)
plt.title('Courbe ROC Régression logistique(cv)', fontsize = 15)
plt.savefig('Graph/Courbe ROC Régression logistique (cv).png', dpi=300, transparent=True)


# Courbe ROC et matrice de confusion Support Vector Machine
matrix_3=metrics.plot_confusion_matrix(clsf_SvM_cv, X, y)
plt.title('Matrice de confusion SVM (cv)', fontsize = 15)
plt.savefig('Graph/Matrice de confusion SVM (cv).png', dpi=300, transparent=True)

roc3=metrics.plot_roc_curve(estimator=clsf_SvM_cv, X=X, y=y)
plt.title('Courbe ROC SVM (cv) ', fontsize = 15)
plt.savefig('Graph/Courbe ROC SVM (cv).png', dpi=300, transparent=True)


# Courbe ROC et matrice de confusion eXtreme Gradient Boost
matrix_4=metrics.plot_confusion_matrix(clsf_XgB_cv, X, y)
plt.title('Matrice de confusion avec XGBoost (cv)', fontsize = 15)
plt.savefig('Graph/Matrice de confusion avec XGBoost (cv).png', dpi=300, transparent=True)

roc4=metrics.plot_roc_curve(estimator=clsf_XgB_cv, X=X, y=y)
plt.title('Courbe ROC XGBoost (cv)', fontsize = 15)
plt.savefig('Graph/Courbe ROC XGBoost (cv).png', dpi=300, transparent=True)

```

<div class ="alert alert-block alert-success"> 

Ici, les deux classifiers avec les meilleurs scores sont : RandomForest et XGBClassifier avec un AUC de 100% pour chacun néanmoins on peut observer une sensibilité qui diffère légèrement au profit de RandomForest.

$ \ $    
    
Le ROC AUC prend en compte les probabilités prédites pour déterminer la performance du modèle. Mais il ne prend en compte que l'ordre des probabilités et ne tient donc pas compte de la capacité du modèle à prédire une probabilité plus élevée pour les échantillons plus susceptibles d'être positifs (Log Loss).

$ \ $  
    
Alors que la métrique AUC précédente est calculée par rapport à une classification binaire avec un seuil de décision variable, la perte logarithmique prend en compte la "certitude" de la classification.

$ \ $  
    
Les résultats jusqu'ici nous renvoient à un choix entre un modèle qui prend compte la capacité à prédire une probabilité plus élevée pour les échantillons plus susceptibles d'être positifs (Log Loss) et des modèles qui ne prennent en compte que l'ordre des probabilités.
 
$ \ $      
    
Aussi, pour les départager on va mettre en lumière ces différentes métriques dans des rapports les résumant avec en plus le score F1 qui est la moyenne harmonique entre la précision et le rappel. Il indique la précision du classificateur (combien d'instances il classe correctement), ainsi que sa robustesse (il ne rate pas un nombre significatif d'instances).Plus le score F1 est élevé, meilleures sont les performances de notre modèle.


    
</div>

Pour le score F1, qui est la moyenne harmonique entre la précision et le rappel, (combien d'instances sont classées correctement et la robustesse, à savoir qu'il ne se trompe pas  sur un nombre significatif d'instances) il apparaît   RandomForest en tête et SVM dernier.

Aussi, eu égard performances évaluées et à l'équilibre du modèle recherché, il est clairement établi que les meilleurs classifiers dans cette modélisation sont Random Forest et la régression Logistique. Néanmoins, XGBoost est préférable car il enregistre le moins de faux négatifs.

<div class ="alert alert-block alert-info"> 
    
Notre nouvel outil de diagnostic basé sur les 6 caractéristiques de noyaux de cellules de tumeurs bénignes et malignes utilisant un algorithme XGBoost est prêt.
    
Aussi, on va donner à notre outil des mesures issues d'un nouveau patient pour prédire s' il est atteint d'un cancer et la probabilité de l'avoir si la tumeur est diagnostiquée bénigne.

    
</div>

```python
# dictionnaire caracteristiques new_patient = {''concavity_mean', 'compactness_mean', 'radius_se', 'concave points_worst', 'radius_worst'} 
new_patient = {'concavity_mean': 0.1440, 'compactness_mean': 0.1213,
               'radius_se': 0.9393,'concave points_worst': 0.06718, 'radius_worst': 14.47,'area_worst': 321 }

# Création du dataset de test X_test qui contient seulement l'echantillon le nouveau patient
X_test = pd.DataFrame([new_patient], index=['new_patient'])
X_test
```

```python
# Liste des caractéristiques dans le panel
panel = X_test.columns 
print(panel)

# Utilisation de la totalité de données disponibles en tant que dataset d'entrainement X_train.
X_train = data_2[panel]
y_train = y
```

```python
# Entrainement de XGbooset et prédiction pour le patient

scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

cls_XGB_fin = XGBClassifier(n_estimators=5, random_state=random_state, eval_metric='error')
cls_XGB_fin.fit(X_train_scaled, y_train)
y_new_patient = cls_XGB_fin.predict(X_test_scaled)
print(cls_XGB_fin, y_new_patient)
```

```python
# resultat du test
test_result=pd.DataFrame(y_new_patient,index=X_test.index,columns=['Type de tumeur'])
test_result
```

<div class ="alert alert-block alert-success"> 
    
Le patient n'a pas de cancer, neanmoins, quelle est la probabilité prédite pour ce patient d'avoir un cancer du sein.Aussi, On va utiliser la méthode `predict_proba()` du modèle à la place de `predict()` pour obtenir la probabilité de prédiction.
    
</div>

```python
# probabilité avec Xgboost
proba_XGB_new_patient = cls_XGB_fin.predict_proba(X_test_scaled)
# Mise en forme du datafram des réultats
proba_result=pd.DataFrame(proba_XGB_new_patient,index=test_result.index, columns=["proba_B","proba_M"])
proba_result
```

<div class ="alert alert-block alert-success"> 

**Résultat:**

Le patient n'a pas de cancer, cependant il a 44 % de risque de developper un cancer du sein. En somme, ce résultat suggère au medecin un suivi plus rapproché malgré le caractère benin de la tumeur du patient.
