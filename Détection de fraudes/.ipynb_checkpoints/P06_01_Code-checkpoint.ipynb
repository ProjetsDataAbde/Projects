**Test d'adéquation à une loi normale**

on va tester l'adéquation à une loi normal qui est un préalable avant d'opérer ACP :

```python
# Test d'adéquation de Kolmogorov-Smirnov si variable 'height_left' suit une loi normale
ks_2samp(data['height_left'],list(np.random.normal(np.mean(data['height_left']), np.std(data['height_left']), 1000)))
```

La p-valeur valant **0.64**,On ne rejette pas l'hypothèse de normalité au niveau de test 5%.   

**Teste de comparaison**

```python
# Constitution des valeurs à comparer
vrai = data.loc[data["is_genuine"]== True]["height_right"]
faux = data.loc[data["is_genuine"]== False]["height_right"]
```

On teste tout d’abord l’égalité des variances avec un test de Bartlett :

```python
# Test égalité des variances
st.bartlett(vrai,faux)
```

La p-valeur valant **0.19**, on rejette l’égalité des variances à un niveau de test de 5%. 

On teste l’égalité des moyennes avec un test de Student :

```python
# Test égalité des moyennes
st.ttest_ind(vrai,faux, equal_var=True)
```

On obtient une p-valeur égale à **6.66**. On constate donc que l’hypothèse d’égalité des moyennes est largement rejetée à un niveau de test de 5%.

**ANOVA :**

On va  mesurer le rapport de corrélation entre la variable <b>is_genuine</b> et la variable <b>diagonal</b> en utilisant le modele <b>ANOVA</b>, puis on evaluera la qualité du model avec eta carré <b>$ η^2 $</b>:

       
$ \ $  


Rapport de corrélation eta carrée $ η^2 $:   
    
$ \displaystyle\ η^2_{Y/X}=\frac {Vinterclasses}{Vtotale}$
   
$ \ $    
    
Si $η^2_{Y/X}=0 $  , alors, cela signifie que les moyennes par classes sont toutes égales,donc a priori pas de relation entre les variables is_genuine et diagonal.
        
$ \ $    
        
Si $η^2_{Y/X}=1 $ ,alors,  cela signifie que les moyennes par classes sont très différentes, chacune des classes étant constituée de valeurs identiques, donc, il existe a priori une relation entre les variables is_genuine et diagonal.

```python
# calcul
X1 = "is_genuine" # qualitative
Y2 = "diagonal" # quantitative

def eta_squared(x,y):
    moyenne_y = y.mean()
    classes = []
    for classe in x.unique():
        yi_classe = y[x==classe]
        classes.append({'ni': len(yi_classe),
                        'moyenne_classe': yi_classe.mean()})
    SCT = sum([(yj-moyenne_y)**2 for yj in y])
    SCE = sum([c['ni']*(c['moyenne_classe']-moyenne_y)**2 for c in classes])
    return SCE/SCT
    
eta_squared(data[X1],data[Y2])
```

 <div class ="alert alert-block alert-success"> 

On a obtenu $η^2_{Y/X}=0,019$, cela signifie que les moyennes sont quasi identique, donc à priori <b>il n'existe pas de correlation entre la variable is_genuine et la variable diagonal</b>. 
    
$ \ $ 
    
A ce stade de l'étude, on a pu constater avec les tests effectués que les variables quantitatives prisent séparement, influent diffèrement sur les valeurs prisent par la variable qualitative dichotomique 'is_genuine'. 
    
Aussi, on va opérer une ACP afin d'avoir une vision globale de toutes les variables, dont l'objectif sera de comprendre et d'approfondir notre connaissance des relations entre variables et individus pour révéler la structure interne des données.
    
</div>

 <div class ="alert alert-block alert-info">

Dans cette partie, on va utiliser l'algorithme de clustering Kmeans de la librairie de Sklearn. 

Cet algorithme évalue la distance entre chaque point et les centroïdes.

En somme, pour chaque point on va calculer la distance euclidienne entre ce point et chacun des centroïdes puis on va l’associer au centroïde le plus proche c’est-à-dire celui avec la plus petite distance. L'objectif ici est de réveler la structure interne en terme de similarité et de dissimilarité sur le premier plan factoriel.
    
</div>

**Clustering** à l'aide du module KMeans:

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import decomposition

# Nombre de clusters souhaités
n_clust = 2
# préparation des données pour le clustering
X = data.values

# Centrer et réduire
X_scaled = preprocessing.StandardScaler().fit_transform(X)

# Clustering par K-means
y = KMeans(n_clusters= n_clust)
y.fit(X)

# Récupération des clusters attribués à chaque individu
clusters = y.labels_

# Affichage du clustering par projection des individus sur le premier plan factoriel
pca = decomposition.PCA(n_components=2).fit(X)
X_projected = pca.transform(X)
plt.scatter(X_projected[:, 0], X_projected[:, 1], c=clusters.astype(np.float), cmap = 'jet', alpha=.2)
plt.title("Projection des {} individus sur le 1e plan factoriel".format(X_projected.shape[0]))
plt.show(block=False)

plt.figure()
centroids = y.cluster_centers_
centroids_projected = pca.transform(centroids)
plt.scatter(centroids_projected[:,0],centroids_projected[:,1])
plt.title("Projection des {} centres sur le 1e plan factoriel".format(len(centroids)))
plt.show()
```

```python
# Distribution des deux classes 
clust_TF = pd.DataFrame(clusters) 
clust_TF.rename(columns = {0:'Clusters'}, inplace = True)
clust_TF=clust_TF.groupby(['Clusters']).size()
clust_TF
```

**Création de la matrice de données (avec les features) et le vecteur des cibles (targets)**

```python
# On recharge les données
data = pd.read_csv("notes.csv", sep =",", header=0)
```

```python
# Création de la matrice de données
random_state = 42
X = data.select_dtypes('number')
X.head()
```

```python
# création du vecteur des cibles
y = data["is_genuine"]
print(y)
```

**Création du dataset d'entrainement et du dataset de test**

On partage le dataset initial `X` en deux datatsets : un dataset qui servira à entrainer le modèle `X_train` et un autre pour tester le modèle obtenu `X_test`. Les échantillons sont tirés au hasard. on va utiliser la librairie **scikit-learn** qui contient l'outil `train_test_split` pour créer ces datasets. 

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=42, stratify=y)
print('Train dataset:', X_train.shape, 'Test dataset:', X_test.shape)
X_train.head(3)
```

**Application d'une normalisation centrée-réduite aux données**

La normalisation centrée-réduite soustrait la moyenne et divise par l'écart-type les valeurs des mesures.

Le calcul de la moyenne $\mu$ et l'écart-type $\sigma$ est réalisé uniquement sur le dataset d'entrainement. Le dataset de test n'est pas utilisé dans le calcul. Il sera normalisé en utilisant les valeurs $\mu$ et $\sigma$ calculées sur le dataset d'entrainement.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler() # crée un scaler
scaler.fit(X_train) # calcule mu et sigma sur X_train uniquement
```

```python
print('Mean mu', scaler.mean_)
print('Std sigma', scaler.scale_)
```

On applique les valeurs $\mu$ et $\sigma$ calculées pour X_train aux deux datasets : **X_train** et **X_test**. Pour cela, on utilise la méthode `transform` du *scaler*. 

```python
# application mu et sigma au datasets x_train et x_test
X_train_scaled = scaler.transform(X_train) # numpy object
# conversion DataFrame
X_train_scaled = pd.DataFrame(X_train_scaled, index=X_train.index, columns=X_train.columns) 

X_test_scaled = scaler.transform(X_test) # numpy object
# conversion DataFrame
X_test_scaled = pd.DataFrame(X_test_scaled, index=X_test.index, columns=X_test.columns)
```

```python
# vérification de la moyenne = 0
X_train_scaled.mean().head()
```

```python
# Vérification de l'ecart type = 1
X_train_scaled.std().head()
```

**Création d'un modèle de régression logistique et calculer la métrique accuracy**

On va utiliser une fonction analytique, appelée fonction logistique ou fonction sigmoïde, qui a une forme caractéristique en S. 

En optimisant les coefficients de cette courbe, elle va nous permettre d'estimer la probabilité pour un billet d'appartenir à la classe True ou ou à la classe False.

```python
# Importation librairie
from sklearn import metrics
# Fonction d'entrainement,test et calcul des métriques
def calculate_accuracy(classifier, X_train, X_test, y_train, y_test):
    classifier.fit(X_train, y_train)   
    y_pred_train = classifier.predict(X_train)
    y_pred_test = classifier.predict(X_test)
    accuracy_train = metrics.accuracy_score(y_train, y_pred_train)
    accuracy_test = metrics.accuracy_score(y_test, y_pred_test)
    print('Train accuracy:', '{:.3f}'.format(accuracy_train), 'Test accuracy:', '{:.3f}'.format(accuracy_test))
    return accuracy_train, accuracy_test, classifier
```

```python
# Importation librairie
from sklearn.linear_model import LogisticRegression,LinearRegression
classifier = LogisticRegression(random_state=random_state, penalty='none')
# train, 
accuracy_train, accuracy_test, trained_classifier = calculate_accuracy(classifier, X_train_scaled, X_test_scaled, y_train, y_test)
```

**Visualisation matrice de confusion :**

```python
# Matrice de confusion
metrics.plot_confusion_matrix(trained_classifier, X_test_scaled, y_test)
plt.title('Matrice de confusion')
```

<div class ="alert alert-block alert-success">

Avec le modèle de regression logistique, on obtient de très bon resultat, sa précision sur le dataset de test est de **98%**.

</div> 

**Evaluer l'impact de chaque mesures dans le modèle de régression logistique :**

On va rechercher les paramètres $\beta$ du modèle obtenu. Il sont dans l'attribut `coef_`. Plus le coefficient $\beta$ est grand (en valeur absolue), plus l'impact de la mesure correspondante est important dans le modèle.

```python
# Affichage des coefficients beta des mesures 
coefficients = pd.DataFrame(trained_classifier.coef_[0], index=X_train_scaled.columns, columns=['beta'])
coefficients
```

```python
# Visualisation de l'impacte des mesures
coefficients = coefficients.sort_values(by='beta')
coefficients.plot.bar(figsize=(15, 3))
plt.title('Impacte des variables sur le modèle')
```

<div class ="alert alert-block alert-success">
    
Il semble que les mesures  **margin_low, margin_up et lenght** sont les variables qui impactent plus le modèle.
</div> 

**Calcule des performances du modèle de régression logistique en utilisant les 3 top features:**

```python
# Identifiez les  meilleurs  mesures qui impactent le plus fortement le modèle régression logistique
n_features = 3
coefficients['abs_beta'] = coefficients['beta'].abs() # calcul des valeurs absolues des beta
coefficients = coefficients.sort_values(by='abs_beta', ascending=False) # tri par valeurs absolues
top_features = list(coefficients.head(n_features).index) # liste de n meilleurs features
print('Top features régression logistique:', top_features)
```

```python
for i in range(len(top_features)):
    selected_features = top_features[0:i+1]
    print(selected_features)
    accuracy_train, accuracy_test, trained_classifier = calculate_accuracy(classifier, X_train_scaled[selected_features], X_test_scaled[selected_features], y_train, y_test)
```
